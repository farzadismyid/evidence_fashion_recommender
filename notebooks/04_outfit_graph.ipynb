```python
from datasets import load_dataset
import pandas as pd

ds = load_dataset(
    "Marqo/polyvore",
    split="data"
)

df = ds.to_pandas()

df["outfit_id"] = (
    df["item_ID"]
    .str.split("_")
    .str[0]
)

df.head()
```

```python
num_items = len(df)
num_outfits = df["outfit_id"].nunique()

print(f"Items: {num_items:,}")
print(f"Outfits: {num_outfits:,}")
```

```python
outfit_sizes = (
    df.groupby("outfit_id")
      .size()
)

outfit_sizes.describe()
```
