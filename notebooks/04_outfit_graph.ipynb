```python
from datasets import load_dataset
import pandas as pd

ds = load_dataset(
    "Marqo/polyvore",
    split="data"
)

df = ds.to_pandas()

df["outfit_id"] = (
    df["item_ID"]
    .str.split("_")
    .str[0]
)

df.head()
```

```python
num_items = len(df)
num_outfits = df["outfit_id"].nunique()

print(f"Items: {num_items:,}")
print(f"Outfits: {num_outfits:,}")
```

```python
outfit_sizes = (
    df.groupby("outfit_id")
      .size()
)

outfit_sizes.describe()
```

```python
from itertools import combinations
from collections import Counter

pair_counter = Counter()

for _, group in df.groupby("outfit_id"):

    categories = (
        group["category"]
        .dropna()
        .unique()
        .tolist()
    )

    for pair in combinations(sorted(categories), 2):
        pair_counter[pair] += 1

pair_df = pd.DataFrame(
    pair_counter.items(),
    columns=["pair", "count"]
)

pair_df = pair_df.sort_values(
    "count",
    ascending=False
)

pair_df.head(30)
```
